```python
# import the package
from pygbm.simulator import GBM_simulator
import matplotlib.pyplot as plt
import numpy as np
```

```python
# set the paramaters
simulator = GBM_simulator(y0 = 1, mu = 2, sigma = 1)
```

```python
# simulate the GBM path using analytic method
t_values, y_values = simulator.simulate_path(T = 1.0, N = 2**7)
simulator.plot_path(t_values, y_values)
```

```python
# simulate the GBM path using milstein method
t_values, y_values = simulator.simulate_path(T = 1.0, N = 2**7, method = "milstein_method")
simulator.plot_path(t_values, y_values)
```

```python
# compare the GBM path simulated using milstein method and the analytic method
simulator.simulate_compare(1.0, 2**8, 79389, ["exact_method", "euler_method"])
```

```python
# simulate the GBM for 10000 times and plot the sample mean
simulator = GBM_simulator(y0 = 1, mu = 2, sigma = 1)
samples = list()
for i in range(10000):
    t_values, y_values = simulator.simulate_path(T = 1.0, N = 2**7, seed = i)
    samples.append(y_values)
    plt.plot(t_values , y_values , color = "grey", alpha = 0.1)
plt.plot(t_values, np.mean(samples, axis = 0), color = "red", label = "sample mean")
plt.xlabel("Time")
plt.ylabel("Y(t)")
plt.legend()
plt.title("Simulated Sample Geometric Brownian Motion Paths")
plt.show()
```
