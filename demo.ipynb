```python
# import the package
from pygbm.simulator import GBM_simulator
```

```python
# set the paramaters
simulator = GBM_simulator(y0 = 1, mu = 2, sigma = 1)
```

```python
# simulate the GBM path using analytic method
t_values, y_values = simulator.simulate_path(T = 1.0, N = 2**7)
simulator.plot_path(t_values, y_values)
```

```python
# simulate the GBM path using milstein method
t_values, y_values = simulator.simulate_path(T = 1.0, N = 2**7, method = "milstein_method")
simulator.plot_path(t_values, y_values)
```

```python
# compare the GBM path simulated using milstein method and the analytic method
simulator.simulate_compare(1.0, 2**8, 79389, ["exact_method", "euler_method"])
```
